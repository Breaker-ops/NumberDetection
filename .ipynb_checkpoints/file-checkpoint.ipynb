```python
import numpy as np
import glob
import matplotlib.pyplot as plt
import librosa
import pandas as pd
from pathlib import Path
import os
```

```python
DATA_DIR = "data_audio"
data = glob.glob(os.path.join(DATA_DIR, "*.wav"))
```

```python
data
```

```python
audio, sr = librosa.load(data[0], sr=None)
```

```python
len(audio), sr, len(audio) / sr
```

```python
plt.plot(audio)
plt.show()
```

```python
spec = librosa.stft(audio)
spec[0][0]
```

```python
librosa.display.specshow(
    librosa.amplitude_to_db(abs(spec)),
    sr=sr,
)
plt.colorbar()
plt.xlabel("time")
plt.ylabel("rate")
plt.show()
```
